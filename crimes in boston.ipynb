```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
crimes=pd.read_csv('crime.csv', sep = ',',encoding = "ISO-8859-1")
```

```python
crimes.head()
```

```python
crimes.columns
```

### Question 1

```python
crime_by_year = crimes.groupby(['YEAR']).size().reset_index()
sorted_crime_by_year= crime_by_year.sort_values([0], ascending = False)
sorted_crime_by_year[:5]
sorted_crime_by_year.rename(columns={0:'Crime count'}, inplace=True)
sorted_crime_by_year.plot.bar(x='YEAR',y='Crime count')
```

### Question 2

```python
crime_by_dist  = crimes.groupby(['DISTRICT']).size().reset_index()
sorted_crime_by_dist = crime_by_dist.sort_values([0], ascending = False)
sorted_crime_by_dist.rename(columns={0:'Crime count'}, inplace=True)
sorted_crime_by_dist.plot.bar(x='DISTRICT',y='Crime count')
```

```python
#Q3  top crime per year
crime_by_year_and_type = crimes.groupby(['YEAR','OFFENSE_CODE_GROUP']).size().reset_index()
sorted_crime_by_year_and_type = crime_by_year_and_type.sort_values(['YEAR',0], ascending = False)
G2015 = sorted_crime_by_year_and_type[sorted_crime_by_year_and_type['YEAR'] == 2015][:5]
G2015.rename(columns={0:'Crime count','OFFENSE_CODE_GROUP':'Crime type'}, inplace=True)
G2015.plot.bar(title='2015',x='Crime type',y='Crime count')
#print(G2015)
G2016 = sorted_crime_by_year_and_type[sorted_crime_by_year_and_type['YEAR'] == 2016][:5]
G2016.rename(columns={0:'Crime count','OFFENSE_CODE_GROUP':'Crime type'}, inplace=True)
G2016.plot.bar(title='2016',x='Crime type',y='Crime count')
#print(G2016)
G2017 = sorted_crime_by_year_and_type[sorted_crime_by_year_and_type['YEAR'] == 2017][:5]
G2017.rename(columns={0:'Crime count','OFFENSE_CODE_GROUP':'Crime type'}, inplace=True)
G2017.plot.bar(title='2017',x='Crime type',y='Crime count')
#print(G2017)
G2018 = sorted_crime_by_year_and_type[sorted_crime_by_year_and_type['YEAR'] == 2018][:5]
#print(G2018)
G2018.rename(columns={0:'Crime count','OFFENSE_CODE_GROUP':'Crime type'}, inplace=True)
G2018.plot.bar(title='2018',x='Crime type',y='Crime count')
```
